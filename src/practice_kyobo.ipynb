```python
import os
import pandas as pd

# 고정된 카테고리 목록
categories = ["fantasy", "history", "horror", "muhyop", "mystery", "romance", "sf"]

# 월별 결과를 저장할 DataFrame 초기화
final_result = pd.DataFrame(columns=categories, index=range(1, 13)).fillna(0)

for month in range(1, 12):  # 1월부터 12월까지 처리
    project_root = os.path.dirname(os.getcwd())  # 상위 디렉터리로 올라가기
    data_dir = os.path.join(project_root, 'data_kyobo')
    file_name = os.path.join(data_dir, f'kyobo_bestseller_{2024}년 {month:02d}월.csv')
    
    # 1. CSV 파일 읽기
    if not os.path.exists(file_name):
        print(f"File not found: {file_name}")
        continue

    df = pd.read_csv(file_name)
    
    # 2. 'Category' 열에서 '일반'이 아닌 값 필터링
    filtered_df = df[df["Category"] != "일반"]
    
    # 3. 월별 데이터 집계
    monthly_counts = (
        filtered_df.groupby("Category")["Category"]
        .count()
        .reindex(categories, fill_value=0)  # 고정된 카테고리를 기준으로 정렬 및 없는 카테고리는 0
    )
    
    # 4. 결과 저장
    final_result.loc[month] = monthly_counts.values

# 5. 결과를 CSV로 저장
output_data_dir = os.path.join(project_root, 'data_analysis')
output_file = os.path.join(output_data_dir, "genre_novel(kyobo).csv")
final_result.to_csv(output_file, index_label="Month", encoding="utf-8-sig")

# 6. 결과 확인
print("카테고리 데이터가 월별로 집계되어 저장되었습니다!")
print(final_result)
```

```python
import pandas as pd
import os

project_root = os.path.dirname(os.getcwd())  
input_dir = os.path.join(project_root, 'data_kyobo')  # 월별 CSV 파일이 저장된 디렉토리
output_file =os.path.join(project_root, 'genre_bestseller_kyobo.csv')  # 최종 저장할 파일 경로

# 필터링된 데이터를 저장할 리스트
filtered_data = []

# 디렉토리 내 모든 CSV 파일 읽기
for filename in os.listdir(input_dir):
    if filename.endswith(".csv"):  # CSV 파일만 처리
        file_path = os.path.join(input_dir, filename)
        
        # CSV 파일 읽기
        data = pd.read_csv(file_path)
        
        # 조건: 'category'가 "일반"이 아닌 데이터 필터링
        filtered = data[data['Category'] != '일반']
        
        # 필터링된 데이터 추가
        filtered_data.append(filtered)

# 모든 필터링된 데이터 병합
final_data = pd.concat(filtered_data, ignore_index=True)

# 최종 데이터 저장
final_data.to_csv(output_file, index=False, encoding='utf-8-sig')  # UTF-8 인코딩

print(f"Filtered data has been saved to {output_file}")
```

```python
import pandas as pd
import os

# 데이터가 저장된 디렉토리 경로
project_root = os.path.dirname(os.getcwd())  
input_dir = os.path.join(project_root, 'data_kyobo_t')  # 월별 CSV 파일이 저장된 디렉토리
output_file =os.path.join(project_root, 'nobel_bestseller_kyobo.csv')  # 최종 저장할 파일 경로

# 병합된 데이터를 저장할 리스트
merged_data = []

# 디렉토리 내 모든 CSV 파일 읽기
for filename in os.listdir(input_dir):
    if filename.endswith(".csv"):  # CSV 파일만 처리
        file_path = os.path.join(input_dir, filename)
        
        # CSV 파일 읽기
        data = pd.read_csv(file_path)
        
        # 데이터 추가
        merged_data.append(data)

# 모든 데이터를 병합
final_data = pd.concat(merged_data, ignore_index=True)

# 최종 데이터 저장
final_data.to_csv(output_file, index=False, encoding='utf-8-sig')  # UTF-8 인코딩

print(f"Merged data has been saved to {output_file}")
```
